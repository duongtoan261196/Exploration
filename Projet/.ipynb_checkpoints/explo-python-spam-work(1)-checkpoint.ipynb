Repésentation par positionnement multidimensionnel

```python
#Author: http://www.nervouscomputer.com/hfs/cmdscale-in-python/

from __future__ import division
 
import numpy as np
 
def cmdscale(D):
    """                                                                                       
    Classical multidimensional scaling (MDS)                                                  
                                                                                               
    Parameters                                                                                
    ----------                                                                                
    D : (n, n) array                                                                          
        Symmetric distance matrix.                                                            
                                                                                               
    Returns                                                                                   
    -------                                                                                   
    Y : (n, p) array                                                                          
        Configuration matrix. Each column represents a dimension. Only the                    
        p dimensions corresponding to positive eigenvalues of B are returned.                 
        Note that each dimension is only determined up to an overall sign,                    
        corresponding to a reflection.                                                        
                                                                                               
    e : (n,) array                                                                            
        Eigenvalues of B.                                                                     
                                                                                               
    """
    # Number of points                                                                        
    n = np.shape(D)[0]
 
    #Etape 1
    A = -0.5 * D**2
    
    #Etape 2: Centering matrix                                                                        
    H = np.eye(n) - np.ones((n, n)).dot(D)/n
 
    #Etape 3
    B = H.dot(A).dot(H.T)
 
    # Diagonalize                                                                             
    evals, evecs = np.linalg.eigh(B)
 
    # Sort by eigenvalue in descending order                                                  
    idx   = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:,idx]
 
    # Compute the coordinates using positive-eigenvalued components only                      
    w, = np.where(evals > 0)
    L  = np.diag(np.sqrt(evals[w])) #racine de delta
    V  = evecs[:,w]
    Y  = V.dot(L)
 
    return Y, evals
```

```python
from sklearn.metrics import euclidean_distances

#distances entre variable
similarities = euclidean_distances(data[liste_rapport + liste_nb_occurrence].T)
dist = 1 - data[liste_rapport + liste_nb_occurrence].corr()**2
```

```python
Y, evals = cmdscale(dist)
```

```python
liste = liste_rapport + liste_nb_occurrence 
```

```python
plt.figure(figsize=(10.,10.))
for i, j, nom in zip(Y[:,0], -Y[:,1], liste):
    plt.text(i, j, nom, color='b')

plt.axis((-0.7,0.7,-0.7,0.7))
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.title('MDS')
plt.show()
```

### Classification des modalités

#### K-means

```python
# Classification d'un grand tableau par k-means puis CAH
## k-means avec 100 classes
from sklearn.cluster import *
k = 100
T=k_means(spam[1:len(spam)],k)
poids=np.zeros((k,))
for i in T[1]:
    poids[i]=1./np.size(T[1])
np.sum(poids)
```

#### CAH

```python
#Definition de la distance pondérée
def weightedL2(a,b,wa,wb):
    q = wa*a-wb*b
    return np.sqrt((q*q).sum())

## Calcul de la matrice des distances pondérées
dist=np.zeros((k,k))
for i in range(dist.shape[0]):
    for j in range(dist.shape[1]):
        dist[i,j]=weightedL2(T[0][i],T[0][j],poids[i],poids[j])
```

```python
##Critères d'agglomération -> ward
from scipy.cluster.hierarchy import *
Q= linkage(dist, 'ward')
```

```python
## Dendrogramme des barycentres des classes
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Q,leaf_font_size=8.)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(maxdists(Q)[::-1][0:10])
plt.title('Hauteur des noeuds')
plt.show()
```

> On retient que $4$ classes

```python
## k-means avec le nombre choisi de classes 
Z=k_means(spam,4)
pca=PCA()
C = pca.fit(spam).transform(spam)
```

```python
#représentation des classes dans l'ACP
plt.figure(figsize=(20, 10))
plt.scatter(C[:,0], C[:,1],marker='1',c=Z[1],cmap=plt.cm.Set1)
plt.title('Classification des individus')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```
