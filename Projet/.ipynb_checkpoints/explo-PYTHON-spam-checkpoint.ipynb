### Analyse en composantes principales

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from scipy.cluster.hierarchy import maxdists
from sklearn import manifold
# définition de la commande
pca = PCA()
```

```python
# Données
Lspam = spam[["spam"] + liste_rapport + liste_nb_occurrence]
Lspam[liste_rapport + liste_nb_occurrence] = np.log(1 + Lspam[liste_rapport + liste_nb_occurrence])
```

```python
#ACP centree non reduit
data_non_reduit = pd.DataFrame(Lspam)
Comp_non_reduit  = pca.fit(data_non_reduit).transform(data_non_reduit)
```

```python
# coordonnées des variables
coord1=pca.components_[0]*np.sqrt(pca.explained_variance_[0])
coord2=pca.components_[1]*np.sqrt(pca.explained_variance_[1])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)
for i, j, nom in zip(coord1,coord2, data_non_reduit.columns):
    plt.text(i, j, nom)
    plt.arrow(0,0,i,j,color='r')
plt.axis((-1.5,1.5,-1.,1.))
plt.show()
```

```python
#ACP contree reduite
data = pd.DataFrame(scale(Lspam),columns = Lspam.columns)
# composantes principales
Comp = pca.fit(data).transform(data)
```

```python
# Eboulis des valeurs propres
plt.figure()
plt.plot(pca.explained_variance_ratio_)
plt.title('Eboulis des valeurs propres')
plt.show()
```

```python
# Diagrammes boîte des composantes principales
plt.figure()
plt.boxplot(Comp[:,0:10])
plt.title('10 premieres composantes principales')
plt.show()
```

```python
# coordonnées des variables
coord1=pca.components_[0]*np.sqrt(pca.explained_variance_[0])
coord2=pca.components_[1]*np.sqrt(pca.explained_variance_[1])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)
for i, j, nom in zip(coord1,coord2, data.columns):
    plt.text(i, j, nom)
    plt.arrow(0,0,i,j,color='r')
plt.axis((-1.2,1.2,-1.2,1.2))
# cercle
c=plt.Circle((0,0), radius=1, color='b', fill=False)
ax.add_patch(c)
plt.show()
```

```python
plt.figure(figsize=(5,5))
for i, j, nom in zip(Comp[:,0], Comp[:,1], Lspam["spam"]):
    color = "red" if nom == 1 else "blue"
    plt.text(i, j, nom, color=color)
plt.axis((-6,30,-10,20))  
plt.show()   
```

```python
plt.plot(Comp[:,0], Comp[:,1], '.')
#plt.axis((-6,7.5,-5,6))
plt.title('Projection des individus')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
plt.show()
```

### Classification des variables

```python
# Stratégie utilisée pour aider à l'interprétation du plan principal 
# lorsque le nombre de variables est important
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(data[liste_rapport + liste_nb_occurrence].T,'ward' ,metric = 'euclidean')
plt.figure(figsize=(15, 8))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Variable')
plt.ylabel('Distance')
dendrogram(Z,leaf_rotation=50.,distance_sort =False,leaf_font_size=8.,labels=data[liste_rapport + liste_nb_occurrence].columns)
plt.show()
```

```python

plt.figure(figsize=(10, 8))
plt.plot(maxdists(Z)[::-1][0:10],'*')
plt.show()
```

```python
from sklearn.metrics import euclidean_distances

similarities = euclidean_distances(data[liste_rapport + liste_nb_occurrence])

mds = manifold.MDS(n_components=2, max_iter=300, eps=1e-9, random_state=None,
                   dissimilarity="precomputed", n_jobs=1)
pos = mds.fit(similarities).embedding_
pos *= np.sqrt((data[liste_rapport + liste_nb_occurrence] ** 2).sum()) / np.sqrt((pos ** 2).sum())

clf = PCA(n_components=2)

pos = clf.fit_transform(pos)

```

```python
fig = plt.figure(1)
ax = plt.axes([0., 0., 1., 1.])

s = 100

plt.scatter(pos[:, 0], pos[:, 1], color='turquoise', s=s, lw=0, label='MDS')
plt.show()
```
